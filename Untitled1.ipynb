```python
pyplot.bar(housing_df['bedrooms'],housing_df['price'])
pyplot.xlabel('Bedrooms')
pyplot.ylabel('Price in Millions')
pyplot.title('Bedrooms Vs Price in Millions')
pyplot.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns


furnishing_counts = housing_df['furnishingstatus'].value_counts()


plt.figure(figsize=(10, 7))
plt.pie(furnishing_counts, labels=furnishing_counts.index, autopct="%1.1f%%")
plt.title("Pie Chart of Furnishing Status")
plt.show()
```

```python
import pandas as pd
from matplotlib import pyplot as plt

# Calculate the distribution of houses on the main road
mainroad_counts = housing_df['mainroad'].value_counts()

# Plot the pie chart
plt.figure(figsize=(8, 6))
plt.pie(mainroad_counts, labels=mainroad_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Houses on Main Road')
plt.show()
```

```python
import pandas as pd
from matplotlib import pyplot as plt



# Aggregate counts for specified columns
columns_of_interest = ['hotwaterheating', 'airconditioning', 'parking', 'bathrooms']
aggregated_counts = pd.Series(dtype=int)

for column in columns_of_interest:
    counts = housing_df[column].value_counts()
    aggregated_counts = aggregated_counts.add(counts, fill_value=0)

# Create a pie chart for the aggregated data
plt.figure(figsize=(10, 8))
plt.pie(aggregated_counts, labels=aggregated_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Housing Features')
plt.show()

```
